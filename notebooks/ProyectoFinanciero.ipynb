```python
import pandas as pd

# 1. Cargamos los datos 
df = pd.read_csv('LINEA_DE_CREDITO_PARA_MYPES_SANTANDER_20260612.csv')

# 2. Estandarizamos los nombres de las columnas (quitamos espacios y pasamos a minúsculas)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# 3. Homologamos la columna 'genero' para que los gráficos queden impecables
# Reemplazamos las siglas por nombres claros 
df['genero'] = df['genero'].replace({
    'F': 'Femenino',
    'M': 'Masculino',
    'PJ': 'Persona Jurídica'
})

# 4. Limpieza de texto general (quitamos espacios rebeldes al inicio o final de los textos)
objeto_columnas = df.select_dtypes(include=['object']).columns
for col in objeto_columnas:
    df[col] = df[col].str.strip()

# 5. Exportamos el resultado a un nuevo archivo CSV LIMPIO
df.to_csv('creditos_santander_limpio.csv', index=False)
print("¡Procesamiento terminado! Se ha generado 'creditos_santander_limpio.csv'")
```
